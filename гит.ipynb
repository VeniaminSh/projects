```python
# находим количество Mr, Mrs, Miss, Master в столбце Name:
total_Mr = 0
total_Mrs = 0
total_Miss = 0
total_Master = 0
for x in train['Name']:
    if "Mr." in x or "Mr " in x:
        total_Mr+=1
    if "Mrs." in x or 'Mrs ' in x:
        total_Mrs +=1
    if "Miss" in x or "Miss." in x:
        total_Miss +=1
    if 'Master' in x or 'Master.' in x:
        total_Master +=1
print(total_Mr + total_Mrs + total_Miss + total_Master)
```

```python
# Находим тех, кто остался в таблице без статуса:
for i in train['Name']:
    if 'Mr.' not in i and "Mrs." not in i and "Miss" not in i and 'Master' not in i and 'Mr ' not in i and 'Mrs ' not in i:
        print(i)
```

```python
# функция для отделения статуса пассажира
def status_name(Name):
    if "Mr." in Name or 'Mr 'in Name:
        return 'Mr'
    if 'Mrs.' in Name or 'Mrs ' in Name:
        return 'Mrs'
    if 'Miss' in Name or 'Miss ' in Name:
        return 'Miss'
    if 'Master' in Name:
        return 'Master'
```

```python
# добавляем новый столбец "status_name" статус людей (холост, женат, замужем, не замужем) 
train.insert(loc=4, column = 'status_name', value = train['Name'].apply(status_name))
train.head(10)
```

```python
#заменим пустые значения стобца "status_name" на unknown
train.loc[train['status_name'].isna(), 'status_name'] = "unknown"

train.isna().sum()
```

```python
# функция для категоризации пассажиров по возрасту (только для тех, по котрым есть данные)
def age_category(age):
    if age < 10:
        return 'up to 10'
    if age < 20:
        return 'up to 20'
    if age < 30:
        return 'up to 30'
    if age < 40:
        return 'up to 40'
    if age < 50:
        return 'up to 50'
    if age < 60:
        return 'up to 60'
    if age < 70:
        return 'up to 70'
    if age <= 80:
        return 'up to 80'
    return 'up to ???'
    
train.insert(loc=7, column = 'age_category', value = train['Age'].apply(age_category))
train.head(20)
```
